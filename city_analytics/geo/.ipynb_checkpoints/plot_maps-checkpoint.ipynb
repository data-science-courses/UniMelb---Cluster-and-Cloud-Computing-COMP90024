# Setup

```python
!pip install cloudant
!pip install geopandas

from pathlib import Path
from cloudant.client import CouchDB
import geopandas as gpd
import folium
from IPython.display import display

# Current working directory
%cd /content/drive/My Drive/unimelb-cluster-and-cloud-computing-comp90024-2020-sm1/city_analytics/geospatial

# Global constants
DATADIR = Path("../geodata")
RESULTDIR = Path("../georesults")
DATAFILE = DATADIR / "aus_sa4_stats_synthetic.json"
MAPFILE = RESULTDIR / "aus_sa4_stats_synthetic.html"
TITLE = "Synthetic statistics in Statistical Areas Level 4, Australia"
```

# Tweet data

```python
client = CouchDB("admin", "password", url='http://172.26.132.48:5984', connect=True)
```

# Load stat-geo data

```python
data = gpd.read_file(DATAFILE)
print(data.info())
display(data.head())
data.plot()
```

# Plot polygons

```python
m = folium.Map(location=[-23.6980, 133.8807], zoom_start=4, tiles="cartodbpositron")

choropleth = folium.Choropleth(
    geo_data=data, data=data,
    columns=["sa4_name", "tweet_count"],
    key_on="feature.properties.sa4_name",
    fill_color="YlOrRd",
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name="Synthetic statistics"
).add_to(m)

choropleth.geojson.add_child(
    folium.features.GeoJsonTooltip(["sa4_name"], labels=False)
)

folium.LayerControl().add_to(m)

# Result
display(m)
m.save(outfile=MAPFILE.as_posix())
```
